### API Demo

```python
covid_api = 'https://api.covidtracking.com/v1/us/daily.json'
```

```python
response_covid_api = requests.get(covid_api)
response_covid_api
```

```python
parsed_json = response_covid_api.json()
parsed_json
```

```python
df_covid = pd.DataFrame(parsed_json)
df_covid
```

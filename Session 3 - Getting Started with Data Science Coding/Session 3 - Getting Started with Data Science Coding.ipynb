## Series

A simple definition of Series: <b>it's a 1-dimensional array</b>.

To create your own Series, you can input <b>pd.Series()</b> with arguments <b>data=[list_of_values]</b>, <b>index=[list_of_index]</b><br>

Take note that the index indicates the position of the row.

In the next cell, you will see an example of a Series with sample grades as values

```python
s_grades = pd.Series(data=[90,85,93,77,88], index=[0,1,2,3,4])
s_grades
```

### Series Familiarization!!
Now, let's get ourselves familiarized with some of the most widely used attributes and callable functions for Series.

#### type(): to check datatype

```python
type(s_grades) 
```

#### len(): to check the size of the Series

```python
len(s_grades)
```

#### .sum(): to get the summation of the values in a Series

```python
s_grades.sum()
```

#### .values: returns the values of the Series <u>in an array format</u>

```python
s_grades.values
```

#### .tolist(): returns the values of the Series <u>in a list format</u>

```python
s_grades.tolist()
```

#### Note: .values and .tolist() are both iterable using a for loop.

```python
# using .values in traditional for loop
for x_values in s_grades.values:
    print(x_values)
```

```python
# using .values in list comprehension
[x_values for x_values in s_grades.values]
```

```python
# using .tolist() 
for x_list in s_grades.tolist():
    print(x_list)
```

```python
# using .tolist() in list comprehension
[x_list for x_list in s_grades.tolist()]
```

### Descriptive Functions

#### mean(): to check the average 

```python
s_grades.mean() # average
```

```python
(s_grades.sum())/len(s_grades) # manual averaging
```

#### min(): to check the lowest value from the Series

```python
s_grades.min()
```

#### max(): to check the highest value from the Series 

```python
s_grades.max()
```

#### describe(): summarize descriptive statistics

```python
s_grades.describe()
```

### Manipulation and Slicing

#### .iloc: locate by index 

```python
s_grades.iloc[0]
```

```python
# you can remove the ".iloc" as a shortcut
s_grades[0]
```

```python
# you can do multiple index selection by specifying it inside a list
s_grades[[0,1,3]] 
```

```python
# you can slice using ':' followed by a number signifying up to what index-1
s_grades[:3] 
```

#### Assigning a new value to a specific index

```python
s_grades[0] = 98
s_grades # 98 is now the new value for index 0
```

### Filtering based on conditions

```python
# filter: where value is less than 85
s_grades[s_grades<85] 
```

```python
 # filter using 2 conditions: where value is greater than 92 and 95
s_grades[(s_grades>92) & (s_grades>95)]
```

### Sample plot using univariate

```python
# univariate
s_grades.plot.barh()
```

## DataFrame

A simple definition of DataFrame: <b>it's a 2-dimensional array.</b>

For our dataset, we used the one we download from Kaggle: https://www.kaggle.com/datasets/residentmario/ramen-ratings?select=ramen-ratings.csv

In the next cell, you will read the csv file using .read_csv()

```python
df = pd.read_csv('ramen-ratings.csv')
```

### DataFrame Familiarization!!

Let's get ourselves familiarized with some of the most widely used attributes and callable functions for DataFrame.

#### .shape: to check the number of rows and columns

```python
df.shape # since the returned value is a tuple (2580, 7), we can access it by index
```

```python
df.shape[0] # number of rows
```

```python
df.shape[1] # number of columns
```

#### Once you open the dataframe, it will provide you the default top 5 and bottom 5 rows

```python
df
```

#### .head(): returns the top 5 rows

```python
df.head()
```

#### .tail(): returns the bottom 5 rows

```python
df.head()
```

#### .columns: returns the list of column available in the dataframe

```python
df.columns
```

#### .dtypes: returns the data type of each column
* object = categorical/nominal values, you can think of this a string literal
* int64/float64 = numerical values or continuous

```python
df.dtypes
```

### Further review in 'Stars' column: check what makes it 'object' dtype

#### .value_counts(): frequency count per value

```python
df['Stars'].value_counts()
```

#### remove the 'Unrated' value and convert it to float

```python
df = df[df['Stars']!='Unrated'] # reassign the filtering to df object again
```

#### .astype(): convert the 'object' dtype into 'float' dtype

```python
df['Stars'] = df['Stars'].astype(float)
```

#### .info(): provides quick summary of the DataFrame containing the column names, its value null count and datatypes

```python
df.info()
```

#### .isnull(): checks the value if it's null or no
#### .sum(): returns the summation of values

<u>Take note</u>: you can specify succeeding callable function or combination of other callable function.

See cell below, for <b>.isnull().sum()</b>:

```python
df.isnull().sum()
```

#### .describe(): returns DataFrame's descriptive statistics 

You can specify the parameter <b>include</b> with an argument of the datatype you want to describe.

```python
# descriptive stats: categorical/nominal 
df.describe(include=['object'])
```

```python
# descriptive stats: all values
df.describe(include='all')
```

#### Nested callables/functions

<u>Remember</u>: you can call multiple callable functions right after each one. Don't forget to always do <shift+tab> on each callable function to explore arguments.

In the next cell, we will use **.plot()** at the end to demonstrate that we can perform visualization based on frequency count distribution using **.value_counts()** 

```python
df.Country.value_counts().sort_values().plot(kind='barh',figsize=[10,10])
```

#### Inplace Argument

Instead of reassigning to the same variable, most of the callables in pandas has **inplace** argument which when set to True, will update the current object based on whatever changes you performed with it.

```python
df.set_index('Review #',inplace=True) # isntead of assigning, add inplace=True as an argument
```

#### .min(): returns the lowest value in a column

#### .max(): returns the highest value in a column

### DataFrame Slicing

#### Column Selection
Remember, you can select multiple columns and put it as a list either directly in the DataFrame or storing it in a variable

```python
df[['Variety','Country','Stars']] # directly in a DataFrame
```

```python
selected_columns = ['Variety','Country','Stars'] # storing it in a variable separately
df[selected_columns]
```

```python
df[selected_columns].head() # performing a callable .head() after the column selection
```

Using the Review #, you can perform row selection. Further, you can also select column/s. 

The cell below utilizes the **.loc** attribute which means you can specify your slicing based on index and column names.

```python
df.loc[2580:2578] # Review# index
```

```python
df.loc[2580:2578, ['Country','Stars']] # Review# index and list of columns
```

### Filtering based on conditions

```python
# filter: where value of Stars > 3.0
df[df['Stars']>3.0] 
```

```python
# filter using 2 conditions: where value of Stars > 3.0 and Country is Philippines
df[(df['Stars']>3.0) & (df['Country']=='Philippines')]
```

```python
# filter: search for keyword using .contains() - i.e., locating 'Canton'
df[df['Variety'].str.contains('Canton')]
```

```python
# filter: search for specific value using '==' comparison operator
df[df['Country']=='Philippines']
```

### DataFrame: Column/Row Drop

#### Removing of rows

Remember, you can reassign DataFrame in a variable or utilize **inplace** argument

```python
df_removed = df.drop(2580) # using index to remove a particular row; then, reassign it
df_removed
```

```python
df.drop(2580,inplace=True) # using index to remove a particular row; then, utilizes inplace argument to true
df
```

#### Removing of columns

**axis=1** means that you are referring to the vertical position in a DataFrame; hence, the column

```python
df.drop(columns='Top Ten',axis=1,inplace=True)
df
```

### DataFrame Groupby Function

You can perform groupby() callable to gain insights about categorical values.

<u>Remember</u>, you can perform multiple callable function right after each other.

```python
# grouped by County; selected Stars to get the average ratings per country
df.groupby('Country')[['Stars']].mean().sort_values(by='Stars')
```

```python
# grouped by County; selected Stars to get the number of ratings per country
df.groupby('Country')[['Stars']].count().sort_values(by='Stars')
```

```python
# grouped by Brand; selected Stars to get the average ratings per Brand
df.groupby('Brand')[['Stars']].mean().sort_values(by='Stars')
```

```python
# grouped by Brand; selected Stars to get the descriptive statistics of ratings per Brand
df.groupby('Brand')[['Stars']].describe()
```
